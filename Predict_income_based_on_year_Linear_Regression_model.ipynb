```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
data = pd.read_csv("canada_per_capita_income.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data.rename(columns={'per capita income (US$)': 'income'}, inplace = True)
data
```

```python
reg_mod = linear_model.LinearRegression()
reg_mod.fit(data[['year']], data.income)
```

```python
reg_mod.predict([[2020]])
```

```python
reg_mod.coef_
```

```python
reg_mod.intercept_
```

```python
828.46507522*2020 + (-1632210.7578554575)
```

```python
plt.scatter(data['year'], data['income'], color = 'r', marker = 'x')
plt.xlabel('year')
plt.ylabel('income in US$)')
plt.plot(data.year, reg_mod.predict(data[['year']]), color='green')
```
